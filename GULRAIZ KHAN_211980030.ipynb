# Pandas Function to get an Overview of the Dataset

## data.head()
#### 'head()' method in Pandas used to display first n rows of a Data Frame or a series. By default it dispays the first 5 rows.

```python
data.head()
```

## data.tail()
#### The 'tail()' mehod is Pandas is the counterpart of the 'head()' method. It is used to display last n rows of the Data Frame or a series. By default display last 5 rows.

```python
data.tail()
```

## data.sample()
#### The 'sample()' method in Pandas is used to randomly sample rows or columns from a Data Frame. 

```python
data.sample(5)
```

## data.info()
#### The 'info()' method in Pandas provides a concise summary of a Data Frame, including information about data types, numbers of non-null values and memory usage.

```python
data.info()
```

## data.describe()
#### The 'describe()' method in Pandas provides descriptive statistics of a Data Frame or series.It computes various summary statistics such as count, mean, standard deviation, minimum, maximum, and numerical columns.

```python
data.describe()
```

## data.query()
#### The 'query()' method in Pandas is used to filter rows from a Data Frame based on specified conditions. The 'query()' method supports both simple and complex condition expressions.

```python
data.query('Quantity == 55')
```

## data.loc()
#### The 'loc()' method in Pandas is used to access rows and columns in a Data Frame by label-based indexing.

```python
data.loc[25 , 'Quantity']
```

## data.unique()
#### The unique() method in Pandas is used to extract the unique values from a column or Series. 

```python
data.nunique(0, True)
```

# Pandas Function to get Deeper Insight into the Datasat

## data.nunique()
#### The 'nunique()' method in Pandas is used to count the number of unique values in a Series or a column of a DataFrame. It returns the count of distinct elements in the original Series or column.

```python
data.nunique()
```

## data.isnull()
#### The 'isnull()' method in Pandas is used to detect missing or null values in a DataFrame or Series. It returns a boolean mask where True indicates a missing value and False indicates a non-missing value.

```python
data.isnull()
```

## data.fillna()
#### The fillna() method in Pandas is used to fill missing or null values in a DataFrame or Series with specified values. It provides several strategies for replacing missing values, such as using a constant value, filling forward or backward, or using statistical methods.

```python
data.fillna(2)
```

## data.short_values()
#### The 'sort_values()' method in pandas is used to sort a DataFrame by one or more columns. The default sort method is in ascending order, placing missing values at the end.

```python
data.sort_values(by='Quantity')
```

## data.value_counts()
#### The 'value_counts()' method in Pandas is used to count the occurrence of unique values in a Series.

```python
data['UnitPrice(USD)'].value_counts()
```

## data.nlargest()
#### The 'nlargest()' method in Pandas is used to retrieve the top N largest values from a Series or a column of a DataFrame. It returns a new Series or DataFrame with the specified number of largest values based on a given criteria.

```python
data.nlargest(10, 'UnitPrice(USD)')
```

## data.nsmallest()
#### The nsmallest() method in Pandas is used to retrieve the bottom N smallest values from a Series or a column of a DataFrame.

```python
data.nsmallest(10, 'Quantity')
```

# Pandas Function to Modify the Dataset


## data.copy()
#### The 'copy()' method in Pandas is used to create a copy of a DataFrame or a Series. It returns a new object that is a deep copy of the original object, including the data and the index.

```python
copied_data = data.copy()
data.loc[0 , 'Status'] = 'Delieverd'
data.head()
```

```python
copied_data.head()
```

## data.rename()
#### The "rename()" method in Pandas is used to rename columns or index labels of a DataFrame. It returns a new DataFrame with the updated column or index labels based on the specified mapping.

```python
data = data.rename(columns = {'OrderDate' : 'Order_Date'})
data
```

## data.where()
#### This 'where()' method replaces the value with NaN or any specified value by the user in a dataframe if the value is false.

```python
data['Shipping_Cost(USD)'].where(data['Shipping_Cost(USD)'] > 50)
```

## data.drop()
#### The 'drop()' method in Pandas is used to remove specified rows or columns from a DataFrame. It returns a new DataFrame with the specified rows or columns removed.

```python
data.drop(0 , axis = 0)
```

```python
data.drop('Quantity' , axis = 1)
```
